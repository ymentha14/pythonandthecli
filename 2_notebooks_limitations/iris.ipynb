```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
# Uncommented plots are not useful
iris_df.plot.scatter(x='sepal width (cm)',y='petal length (cm)')
```

```python
iris_df = iris_df[(iris_df['sepal width (cm)'] > 2.5) | (iris_df['petal length (cm)'] > 2.0)]
```

```python
# Code redundancy
iris_df.plot.scatter(x='sepal width (cm)',y='petal length (cm)')
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
y = iris.target

# One should separate computation and plotting functions
X_reduced = PCA(n_components=3).fit_transform(iris.data)


ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```
